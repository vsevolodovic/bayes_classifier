# The Model For Calculation of Discriminant Function Coefficients

```python
def trainBayes(dataset):
    boolean = dataset.y < 1
    df_class0=dataset[boolean].iloc[:,:2]
    boolean = dataset.y > 0 
    df_class1=dataset[boolean].iloc[:,:2]
    
    mu_class0 = np.mean(df_class0, axis=0)
    mu_class0T = np.transpose(np.mean(df_class0, axis=0))
    cov_class0 =df_class0.cov()
    inv_cov_class0 = np.linalg.inv(df_class0.cov())
    lnP_class0 = np.log(len(df_class0)/(len(df_class0) + len(df_class1)))
    ln_det_cov_class0 = np.linalg.det(df_class0.cov())
    
    mu_class1 = np.mean(df_class1, axis=0)
    mu_class1T = np.transpose(np.mean(df_class1, axis=0))
    cov_class1 =df_class1.cov()
    inv_cov_class1 = np.linalg.inv(df_class1.cov())
    lnP_class1 = np.log(len(df_class1)/(len(df_class0) + len(df_class1)))
    ln_det_cov_class1 = np.linalg.det(df_class1.cov())
    
    if np.array_equal(cov_class0,cov_class1) :
        W0t = np.dot(mu_class0T, inv_cov_class0)
        W0  = lnP_class0 - 0.5 * np.dot(W0t, mu_class0)   
        W1t = np.dot(mu_class1T, inv_cov_class1)
        W1  = lnP_class1 - 0.5 * np.dot(W1t, mu_class1)
        return W0t,W0,W1t,W1
   
    else: #general case
        W00 = -0.5 * inv_cov_class0
        W0t = np.dot(mu_class0T,inv_cov_class0)
        W0  = lnP_class0 - 0.5 * np.dot(W0t, mu_class0) - 0.5 * ln_det_cov_class0
        W01 = -0.5 * inv_cov_class1
        W1t = np.dot(mu_class1T,inv_cov_class1)
        W1  = lnP_class1 - 0.5 * np.dot(W1t, mu_class1) - 0.5 * ln_det_cov_class1
        return W00,W0t,W0,W01,W1t,W1
        
        
        
        
       
```

# Training Model With train1.csv

# In this part, the covariance matrices of classes in "train1.csv" seems equal. However, their precision is not equal. Because of this our model considers them as different matrices and solves the problem for general case.

```python
W00,W0t,W0,W01,W1t,W1 = trainBayes(df1) 
```

# Testing Model with "test1.csv" , Calculating Discriminant Function and Calculating Classification Error Rate

```python

testSamples = test1.iloc[:, :-1]
testLabels  = test1.iloc[:, -1]
error_counter1 = 0
i = 0
y_pred = np.zeros(len(test1))
for i in range(len(test1)):
    g_0X = np.dot(np.dot(np.transpose(testSamples.iloc[i, :]),W00), testSamples.iloc[i, :] ) + np.dot(W0t,testSamples.iloc[i, :]) + W0
    g_1X = np.dot(np.dot(np.transpose(testSamples.iloc[i, :]),W01), testSamples.iloc[i, :] ) + np.dot(W1t,testSamples.iloc[i, :]) + W1
    if g_0X > g_1X :
        y_pred[i] = 0
    
    else:
        y_pred[i] = 1

    if y_pred[i] != testLabels[i]:
        error_counter1 += 1

test1_error = (error_counter1/len(test1)) * 100
print("Classification Error Rate of test1.csv :",test1_error,"%")     
```

```python
df1_test1 = pd.concat([df1,test1],axis=0)
features1 = df1_test1.iloc[:,:-1]
labels1 = df1_test1.iloc[:,-1]

from sklearn.model_selection import KFold

kf = KFold(n_splits=3, random_state=1773, shuffle=True)
for train_index, test_index in kf.split(features1):
    
    
    features1_train, features1_test = features1.iloc[train_index,:], features1.iloc[test_index,:]
    labels1_train, labels1_test = labels1.iloc[train_index], labels1.iloc[test_index]
    
    #Dataset for trainBayes function
    train_samples1 = pd.concat([features1_train,labels1_train],axis=1) #Column-wise concatenation
    print(train_samples1.head())
    
    boolean = train_samples1.iloc[:, -1] < 1
    train_samples1_class0=train_samples1[boolean].iloc[:,:-1]
    train_samples1_class0_cov = train_samples1_class0.cov()
    train_samples1_class0_mu = np.mean (train_samples1_class0, axis = 0)
    
    
    boolean = train_samples1.iloc[:, -1] > 0
    train_samples1_class1=train_samples1[boolean].iloc[:,:-1]
    train_samples1_class1_cov = train_samples1_class1.cov()
    train_samples1_class1_mu = np.mean (train_samples1_class1, axis = 0)
    
    #Training Model
    #Because of precision inequality of covariance matrices the model solves the problem for general case
    W00,W0t,W0,W01,W1t,W1 = trainBayes(train_samples1)
    
    
    #Testing Model, Calculating discriminant functions and Calculating Classification Error
    error_counter = 0
    i = 0
    y_pred = np.zeros(len(features1_test))
    for i in range(len(features1_test)):
        g_0X = np.dot(np.dot(np.transpose(features1_test.iloc[i, :]),W00), features1_test.iloc[i, :] ) + np.dot(W0t,features1_test.iloc[i, :]) + W0
        g_1X = np.dot(np.dot(np.transpose(features1_test.iloc[i, :]),W01), features1_test.iloc[i, :] ) + np.dot(W1t,features1_test.iloc[i, :]) + W1
        if g_0X > g_1X :
            y_pred[i] = 0
    
        else:
            y_pred[i] = 1
    labels1_test=labels1_test.to_frame().reset_index()
    for i in range(len(features1_test)):
        
        if(y_pred[i] != labels1_test.iloc[i,1]):
            error_counter+=1

    
    print("Error Rate: ",(error_counter/len(features1_test))*100,"%")


         

    
    
```
